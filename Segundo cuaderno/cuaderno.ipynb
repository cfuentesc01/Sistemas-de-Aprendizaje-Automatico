## 2. Medidas de tendencia central

Calcula la media, mediana y moda

Indica si existe moda real

```python
import numpy as np
import statistics as stats

tiempos = [1.2, 1.5, 1.7, 2.0, 1.3, 1.4, 1.6, 2.1]

print("Esta es la media: ", np.mean(tiempos))
print("Esta es la mediana: ", np.median(tiempos))
print("Esta es la moda: ", stats.multimode(tiempos))

# Realmente no hay moda ya que no hay valor que se repita mas de una sola vez
```

## 3. Cuartiles y rango intercuarílico

```python
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt

visualizaciones = [120, 130, 150, 160, 200, 210, 300, 320]

# Calcula los cuartiles
q1= np.percentile(visualizaciones, 25)
q2= np.percentile(visualizaciones, 50)
q3= np.percentile(visualizaciones, 75)

print("El primer cuartil es: ", q1)
print("El segundo cuartil es: ", q2)
print("El tercer cuartil es: ", q3)

# Calcula rangos intercuartilicos
iqr = q3 - q1
print("El rango intercuartílico es: ", iqr)

# Detecta outliers
outlier_menor = q1 - 1.5*iqr
outlier_mayor = q3 + 1.5*iqr

for num in visualizaciones:
    if num < outlier_menor:
        print("Los outliers superiores son: ", num)
    if num > outlier_mayor:
        print("Los outliers superiores son: ", num)
    else:
        None

# Representa el diagrama boxplt
plt.boxplot(visualizaciones)
plt.title("Diagrama boxplt")
    
```

## 4. Comparación A/B

```python
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt

A = [30, 35, 50, 55, 40, 45, 60] 
B = [20, 25, 30, 35, 28, 32, 34]

# Calcula la media
media_a = np.mean(A)
media_b = np.mean(B)

print("La media de A es : ", media_a)
print("La media de B es : ", media_b)

# Calcula la desviacion típica
desviacion_a = np.std(A)
desviacion_b = np.std(B)

print("La desviacion de A es: ", desviacion_a)
print("La desviacion de B es: ", desviacion_b)

# Calcula el CV
coeficiente_variacion_a = (desviacion_a / media_a) * 100
coeficiente_variacion_b = (desviacion_b / media_b) * 100

print("El CV de A es: ", coeficiente_variacion_a)
print("El CV de B es: ", coeficiente_variacion_b)

# Representa gráficamente
plt.plot(A)
plt.plot(B)
plt.title("Representación de A y B")
```

## 6. Estudio de outliers

```python
import numpy as np

edades = [25, 26, 27, 25, 28, 29, 120]

# Para calcular los outliers, se necesita el rango intercuartilico
edades_mediana = np.median(edades)

# Calcula los cuartiles
q1_edades = np.percentile(edades, 25)
q2_edades = np.percentile(edades, 50)
q3_edades = np.percentile(edades, 75)

print("El primer cuartil es: ", q1_edades)
print("El segundo cuartil es: ", q2_edades)
print("El tercer cuartil es: ", q3_edades)

# Calcula rangos intercuartilicos
iqr_edades = q3_edades - q1_edades
print("El rango intercuartílico es: ", iqr_edades)

# Detecta outliers y eliminarlos
outlier_menor_edades = q1_edades - 1.5*iqr_edades
outlier_mayor_edades = q3_edades + 1.5*iqr_edades

for num in edades:
    if num < outlier_menor_edades:
        print("Los outliers superiores son: ", num)
    if num > outlier_mayor_edades:
        print("Los outliers superiores son: ", num)
        edades.remove(num)
        edades.append(int(edades_mediana))
    else:
        None

print("La nueva lista acaba de esta forma: ", edades)
```

## 7. Probabilidad condicionada

Dadas las siguientes probabilidades: 
- P(móvil)=0.40
- P(tablet)=0.30
- P(PC)=0.30
- P(corto|móvil)=0.80
- P(corto|tablet)=0.50
- P(corto|PC)=0.20

Calcula P(corto) y P(movil|corto)

```python
# Para calcular P(corto)
x = (0.80)*(0.40) + (0.50)*(0.30) + (0.20)*(0.30)
print("La probabilidad de P(corto) es: ", x)

# Para calcular P(movil|corto), deberemos de hacer lo siguiente:
y = (0.40 * 0.80) / (0.40 * 0.80 + 0.30 * 0.50 + 0.30 * 0.20)
print("La probabilidad de P(movil|corto) es: ", y)
```

## 8. Teorema de Bayes

Dadas las siguientes probalidades
- P(fraude)=0.02
- P(alarma|fraude)=0.95
- P(alarma|no fraude)=0.05.

```python
# Tambien necesitaremos P(alarma)
g = 0.95 * 0.02 + 0.05 * 0.98 


# Calcula P(fraude|alarma)
z = (0.02 * 0.95) / (0.02 * 0.95 + 0.05 * g)
print("La probabilidad de P(fraude|alarma) es: ", z)
```

## 9. Regresión Lineal

```python
from sklearn.linear_model import LinearRegression

publicidad = [[1000], [1500], [2000], [2500]]
ventas = [120, 150, 170, 200]

# Ajusta una regresión lineal
modelo = LinearRegression()
modelo.fit(publicidad, ventas)

pendiente = modelo.coef_[0]
independiente = modelo.intercept_

print("La pendiente es: ", pendiente)
print("La variable independiente es: ", independiente)

# Predice para 3000 euros
prediccion_3000 = modelo.predict([[3000]])
print("La predicción de publicidad para 3000 euros es: ", float(prediccion_3000[0]))
```
