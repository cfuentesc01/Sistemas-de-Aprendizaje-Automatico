```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

datos = [5, 7, 8, 8, 9, 10, 10, 14, 14, 15, 20]

print("Media:", np.mean(datos))
print("Mediana:", np.median(datos))
print("Varianza:", np.var(datos))
print("Cuartiles:", np.percentile(datos,[25, 50, 75]))
plt.boxplot(datos)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df_fire = pd.read_csv("fires-all.csv")
df_fire.drop(columns=["lat", "lng"])
#print(df_fire)

media_superficie = df_fire["superficie"].mean()
print("La media es : " ,media_superficie)
mediana_superficie = df_fire["superficie"].median()
print("La mediana es : ",mediana_superficie)
moda_superficie = df_fire["superficie"].mode()
print("La moda es: " ,moda_superficie)

cuartiles = np.percentile(df_fire["superficie"],[25, 50, 75])
print("Los cuartiles son :" , cuartiles)

# Gr√°fica superficie
plt.plot(df_fire["superficie"])
plt.plot(df_fire["idprovincia"])
```
